(2) Describe the meaning of 'stratified split' and the difference between when using stratify=True and stratify=False in function train_test_split() 

In machine learning, splitting the dataset into a training set and a test set is a common approach to evaluate the performance of the model. Stratified splitting is a technique that ensures that the class distribution of the target variable is maintained in both the training and test sets.

When using train_test_split() with stratify=True, the data is split in such a way that the proportion of each target class is preserved in both the training and test sets. This is useful when the target variable has an imbalanced distribution, meaning that some classes have significantly fewer samples than others. By preserving the class distribution in both sets, stratified splitting can prevent bias in the model and improve its accuracy.

On the other hand, when using train_test_split() with stratify=False, the data is split randomly, without considering the target variable. This can be useful in situations where the target variable is not relevant to the model, or if the class distribution is already balanced in the dataset. However, if the target variable has an imbalanced distribution, using stratify=False can lead to a biased model, especially if the number of samples is limited.

**(Bonus) Task 11: Evaluate Bias-Variance decomposition for the classification models**

```python
import numpy as np
from sklearn.metrics import mean_squared_error

def bias_variance_decomp(estimator, X_train, y_train, X_test, y_test):
    # Fit the estimator on the training set
    estimator.fit(X_train, y_train)
    
    # Make predictions on the training and test sets
    y_train_pred = estimator.predict(X_train)
    y_test_pred = estimator.predict(X_test)
    
    # Compute the bias and variance
    bias = np.mean((y_test - np.mean(y_test_pred)) ** 2)
    variance = np.mean((y_test_pred - np.mean(y_test_pred)) ** 2)
    
    # Compute the MSE
    train_error = mean_squared_error(y_train, y_train_pred)
    test_error = mean_squared_error(y_test, y_test_pred)
    
    # Print the results
    print("Bias: ", bias)
    print("Variance: ", variance)
    print("Train Error: ", train_error)
    print("Test Error: ", test_error)
```

```python
# Evaluate bias-variance decomposition for the KNN model
print("KNN Bias-Variance Decomposition:")
bias_variance_decomp(knn_model, X_train, y_train, X_test, y_test)

# Evaluate bias-variance decomposition for the Naive Bayes model
print("Naive Bayes Bias-Variance Decomposition:")
bias_variance_decomp(nb_model, X_train, y_train, X_test, y_test)

# Evaluate bias-variance decomposition for the Decision Tree model
print("Decision Tree Bias-Variance Decomposition:")
bias_variance_decomp(dt_model, X_train, y_train, X_test, y_test)

# Evaluate bias-variance decomposition for the Quadratic Discriminant Analysis model
print("QDA Bias-Variance Decomposition:")
bias_variance_decomp(qda_model, X_train, y_train, X_test, y_test)

# Evaluate bias-variance decomposition for the Support Vector Machine model
print("SVM Bias-Variance Decomposition:")
bias_variance_decomp(svc_model, X_train, y_train, X_test, y_test)

# Evaluate bias-variance decomposition for the Logistic Regression model
print("Logistic Regression Bias-Variance Decomposition:")
bias_variance_decomp(logreg_model, X_train, y_train, X_test, y_test)

# Evaluate bias-variance decomposition for the Linear Discriminant Analysis model
print("LDA Bias-Variance Decomposition:")
bias_variance_decomp(lda_model, X_train, y_train, X_test, y_test)
```
